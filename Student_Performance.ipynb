# Student Study Performance

This project helps on understanding how the student's performance is affected by other variables such as Gender, Ethnicity, Parental level of education and Test preparation course through different data visualization.
     
  girl-998988_640.jpg

```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('study_performance.csv')
```

```python
# Displaying first 5 rows of the dataset
print("\nFirst 5 rows of the dataset:")
print(data.head(5))

# Checking for missing values
print("\nChecking for missing values:")
print(data.isnull().sum())

#no need for data cleaning as no any missing values
```

```python
# Data Exploration
print("Data Exploration:")
print(data.describe())
```

```python
#Visualization
# 1: Gender distribution using countplot
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
print("Insight: There are slightly more females than males in the dataset.")
```

```python
#2:Ethnicity distribution using pie
ethnicity_counts = data['race_ethnicity'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(ethnicity_counts, labels=ethnicity_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Ethnicity Distribution')
plt.axis('equal')
plt.show()
```

```python
print("Insight: Majority of the students belong to group C ethnicity.")
```

```python
# 3: score distribution by gender using box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='gender', y='reading_score')
plt.title('Reading Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()
```

```python
print("Insight: Female have more reading score compared to male.")
```

```python
# 4: test preparation course using countplot
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='test_preparation_course')
plt.title('Test Preparation Course Completion')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.show()
```

```python
print("Insight: A significant number of students have not completed the test preparation.")
```

```python
#5: correlation scores using heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data[['math_score', 'reading_score', 'writing_score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Scores')
plt.show()
```

```python
print("Insight: There is a strong correlation between reading and writing scores, while the correlation between math and reading/writing scores is moderate.")
```
